# VERIFICANDO DISTRIBUIÇÃO
Verificando se a distribução dos tempos é gaussiana ou não possui padrão definido. Sempre fazemos isso usando um histograma, onde primeiro precisamos saber quantas caixas tem o histograma

```python
def numberBins(data):
    n = len(data)
    k = np.log2(n) + 1
    return int(np.ceil(k))
```

### Histograma dos Tempos de Sinal Vermelho

```python
plt.figure(figsize=(7, 4))
k = numberBins(temposVermelho)

sns.histplot(temposVermelho, color='blue', kde=True, bins=k)
plt.show()
```

### Histograma dos Tempos de Sinal Não Vermelho

```python
plt.figure(figsize=(7, 4))
k = numberBins(temposResto)

sns.histplot(temposResto, color='blue', kde=True, bins=k)
plt.show()
```

# INCERTEZA RELATIVA DA MÉDIA

Tabela t-student para precisão e exatidão dos dados por incerteza relativa

```python
tStudent = [
    12.706, 4.303, 3.182, 2.776, 2.571, 2.447, 2.265, 2.306, 2.262, 2.228, 2.201, 2.179,
    2.160, 2.145, 2.131, 2.120, 2.110, 2.101, 2.093, 2.086, 2.080, 2.074, 2.069, 2.064,
    2.060, 2.056, 2.052, 2.048, 2.045, 2.042
]
```

Calculando a incerteza relativa em uma lista de dados com relação a sua média

$I_R = \frac{I_\mu}{\mu} = \frac{T \cdot \frac{\sigma_a}{\sqrt{n}}}{\mu}$

```python
def incertezaRelativa(array):
    global tStudent
    n = len(array)
    T = 1

    if n <= 30:
        T = tStudent[n-2]  

    return T * np.std(array, ddof = 1)/(n**0.5)/np.mean(array)
```

```python
print('TemposVermelho: ', round(np.mean(temposVermelho), 2))
print('TemposResto   : ', round(np.mean(temposResto), 2))
```

```python
print(f'{incertezaRelativa(temposVermelho) * 100:.2f} %')
print(f'{incertezaRelativa(temposResto)    * 100:.2f} %')
```

# ELIMINANDO OUTLIERS

```python
def treatData(list):
    array = np.array(list)
    
    std  = np.std(array)
    mean = np.mean(array)

    upper = mean + 1*std
    lower = mean - 1*std

    return array[(array > lower) & (array < upper)]

temposVermelho = treatData(temposVermelho)
temposResto    = treatData(temposResto)
```

```python
print('TemposVermelho: ', round(np.mean(temposVermelho), 2))
print('TemposResto   : ', round(np.mean(temposResto), 2))
```

```python
print(f'{incertezaRelativa(temposVermelho) * 100:.2f} %')
print(f'{incertezaRelativa(temposResto)    * 100:.2f} %')
```
